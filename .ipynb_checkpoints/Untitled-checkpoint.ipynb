```python
#Les packages 
#Installons les packages necessaires
import pandas as pd # Pandas est utilisé pour analyser les données.
import numpy as np #NumPy est une bibliothèque Python utilisée pour travailler avec des tableaux.
import seaborn as sns #Seaborn est une bibliothèque qui utilise Matplotlib en dessous pour tracer des graphiques.-
#Il sera utilisé pour visualiser des distributions aléatoires.
import matplotlib.pyplot as plt
plt.rcParams ['figure.figsize'] = [6, 2]
```

```python
##Les données
#La méthode choice() vous permet de générer une valeur aléatoire basée sur un tableau de valeurs.
#La méthode choice() prend un tableau comme paramètre et renvoie aléatoirement l’une des valeurs.

#Créons les données aléatoiorement 
dict = { 
    #le  module "random" pour travailler avec des nombres aléatoires.
    #les variables qualitatives 
    "Genre": np.random.choice(["Homme", "Femme","No-binaire"],50),
    "Niveau_educatives" : np.random.choice(["Secondaire","Universitaire","Maitrise","Doctorat"],50),
    "Emploi" : np.random.choice(["Technologie","Finance","Santé","Vente","Education"],50),
    "Statut_emploi" : np.random.choice(["Temps plein","Temps partiel","Contractuel"],50),
    "Zone_residence" : np.random.choice(["Urbain","Rural","Suburbain"],50),
    "Social" : np.random.choice(["Assurance maladie","Régime de retraite","Congé payés"],50),
    "Promotion" : np.random.choice(["Elevées","Moyennes","Faibles"],50),
    
    #Variables quantitatives
    #La méthode randint() prend un size paramètre dans lequel vous pouvez spécifier la forme d'un tableau.
    "Age" :np.random.randint(17,60,50),
    "Salaire" :np.random.randint(1,10,50),
   "Experience" :np.random.randint(1,30,50),
    "Heures_semaine" :np.random.randint(6,35,50)

    
}
```

```python
df = pd.DataFrame(dict)
df.head()
```

```python
#Nuage de Point
sns.scatterplot(x = "Experience", y = "Salaire", data = df, hue = "Genre")
```

```python
#Graphique linéaire
sns.lineplot(x = "Experience", y = "Salaire", data = df)
```

```python
#Courbe sz Sensité 
sns.kdeplot(data = df)
```

```python
#Diagramme en barres
sns.barplot (x ="Genre", y="Salaire", data = df, hue = "Genre")
```

```python
# Counts plot
sns.countplot ( y = "Salaire", data = df,hue ="Zone_residence")
```

```python
#Histogramme
sns.histplot (x="Age", data = df)
```
